# Indexing & Selection

| Feature             |  `.loc`                                         |  `.iloc`                                                    |
| --------------------- | ----------------------------------------------- | ----------------------------------------------------------- |
| **Syntax** | `df.loc[row_indexer, column_indexer]`           | `df.iloc[row_position, column_position]`                      |
| **Indexing Method** | Label-based indexing | Position-based indexing |
| **Used for Reference**  | Row and column labels (names)             | Numerical indices of rows and columns (starting from 0)             |


## Label‑based (`.loc`)


```python
# Viwes row and column labels (names)
df.axes
```

```python
df.head()
```

```python
row2 = df.loc[2]
print(row2)
```

```python
# Select rows 0–3 and columns 'longitude' & 'latitude'
# fixed position [row, column]
subset = df.loc[0:3, ['longitude', 'latitude']]
subset.head()
```

## Position‑based (`.iloc`)

```python
df.axes
```

```python
# find description
# help(df.iloc)
# df.iloc? -? ctrl + enter
```

```python
# Select the first row
first_row = df.iloc[0]
first_row.head()
```

```python
# Select first 3 rows and first 2 columns
subset = df.iloc[0:3, 0:2]
subset.head()
```

# Basic Operations



## Filtering rows

```python
population_more1000 = df[df['population'] > 1000]
population_more1000
```

## Adding & dropping columns



```python
# Add a new column
df['rooms_per_household'] = df['total_rooms'] / df['households']
df.head()
```

```python
# Drop a column
df = df.drop(columns=['rooms_per_household'])
df.head()
```

### Sorting


```python
# Sort by population descending (ascending=False)
df.sort_values(by='population', ascending=False)
```

# Handling Missing Data

```python
# check null with find datatype
df.info()
```

```python
# Detect missing values
df.isnull().sum()
```

```python
# Drop rows with any missing values
df_clean = df.dropna()
df_clean.isnull().sum()
```

```python
# Fill missing values with a default
df_filled = df.fillna({'total_bedrooms': df['total_bedrooms'].mean()})
df_filled.isnull().sum()
```
